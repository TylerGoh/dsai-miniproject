```python
Customers = pd.read_csv('./data/olist_customers_dataset.csv')
Geolocation = pd.read_csv('./data/olist_geolocation_dataset.csv')
Orders = pd.read_csv('./data/olist_orders_dataset.csv')
OrderItems = pd.read_csv('./data/olist_order_items_dataset.csv')
OrderPayments = pd.read_csv('./data/olist_order_payments_dataset.csv')
OrderReviews = pd.read_csv('./data/olist_order_reviews_dataset.csv')
Products = pd.read_csv('./data/olist_products_dataset.csv')
Sellers = pd.read_csv('./data/olist_sellers_dataset.csv')
```

```python
Customers.dtypes
```

```python
Geolocation.dtypes
```

### Customers and Geolocation have different columns names for zip_code which must be renamed

```python
Customers.rename(columns={"customer_zip_code_prefix": "zip_code"}
                 , inplace=True)
Geolocation.rename(columns={"geolocation_zip_code_prefix": "zip_code"}
                   , inplace=True)
```

```python
Geolocation.head()
```

### One zip code can point to 2 different lattitude and longtitude which is pointless hence we remove duplicates

```python
Geolocation.drop_duplicates(subset=['zip_code'], inplace=True)
```

### Merging all dataset based on database schema above

```python
train_data = Customers.merge(Orders, on="customer_id")
train_data = train_data.merge(OrderReviews, on="order_id")
train_data = train_data.merge(OrderItems, on="order_id")
train_data = train_data.merge(OrderPayments, on="order_id")
train_data = train_data.merge(Products, on="product_id")
train_data = train_data.merge(Sellers, on="seller_id")
train_data = train_data.merge(Geolocation, on="zip_code", how="left")
```

```python
import time
def sentimentFunc(row):
    tokens = tokenizer([row['review_comment_message']], return_tensors="pt",
                       padding=True, truncation=True, max_length=512)
    finbertptbr_outputs = finbertptbr(**tokens)
    arr = finbertptbr_outputs.logits.cpu().detach().numpy()
    Positive = arr[0][0]
    Neutral = arr[0][1]
    Negative = arr[0][2]
    return pd.Series([Positive, Neutral, Negative])

df = cleaned_data.copy()
df[['Positive', 'Neutral', 'Negative']] = df.apply(sentimentFunc, axis=1) 
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split


training_data = cleaned_data.copy()

cols_to_remove = []

for col in training_data.columns:
    try:
        _ = training_data[col].astype(float)
    except ValueError:
        cols_to_remove.append(col)
        pass

# keep only the columns in df that do not contain string
training_data = training_data[[col for col in training_data.columns if col not in cols_to_remove]]

y_data = training_data["review_score"]
x_data = training_data.drop(columns=['review_score']);
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree




tree1 = DecisionTreeClassifier(max_depth= 10)    #* tree1 : max depth 2
tree1.fit(x_train, y_train)
f = plt.figure(figsize=(12,12))
plot_tree(tree1, filled=True, rounded=True)
print("-")
```
